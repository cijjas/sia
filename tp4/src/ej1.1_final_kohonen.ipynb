## DISTANCE AVERAGES
Using Detailed Configuration so the analysis is much clearer :))

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import json
from core.kohonen import Kohonen

config_file = "../configs/kohonen_detailed.json"

with open(config_file) as f:
    config = json.load(f)

europe_data = pd.read_csv(config['data']['source'])
df = europe_data[config['data']['features']]
scaler = StandardScaler()
data = scaler.fit_transform(df.values)

som = Kohonen(
    data=data,
    grid_size=config['kohonen']['grid_size'],
    learning_rate=config['kohonen']['learning_rate'],
    eta_function=config['kohonen']['eta_function'],
    radius=config['kohonen']['initial_radius'],
    radius_function=config['kohonen']['radius_function'],
    similarity_function=config['kohonen']['similarity_function'],
    seed=config['kohonen']['seed'],
    weights=None,
    init_method='random_choice'
)

som.fit(num_iterations=config['kohonen']['number_of_iterations'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_umatrix(u_matrix):
    plt.figure(figsize=(8, 8))
    plt.imshow(u_matrix, cmap='viridis', origin='lower')  # Specify color scheme for the heatmap
    plt.colorbar(label='Average Distance to Neighbors')
    plt.title('Unified Distance Matrix', fontsize=14)

    # Set the ticks to start from 1 and increment by 1 for each position
    num_ticks = u_matrix.shape[0]
    tick_positions = np.arange(num_ticks)
    tick_labels = tick_positions + 1  # Labels from 1 to num_ticks

    plt.xticks(tick_positions, tick_labels)
    plt.yticks(tick_positions, tick_labels)

    # Label the x and y axes
    plt.xlabel("X Grid Position")
    plt.ylabel("Y Grid Position")

    plt.show()

# Assuming u_matrix is calculated from the som
u_matrix = som.calculate_umatrix()
plot_umatrix(u_matrix)

def stretch_grid_positions_smooth(grid_size, u_matrix):
    # Initialize positions in a grid
    x, y = np.meshgrid(np.arange(grid_size), np.arange(grid_size))
    x = x.astype(float)
    y = y.astype(float)

    # Center the grid positions
    x_center = (grid_size - 1) / 2
    y_center = (grid_size - 1) / 2

    # Smooth stretching: Apply average of U-matrix values for better stretching
    displacement_factor = 0.1  # Control how much stretching occurs

    for i in range(grid_size):
        for j in range(grid_size):
            # Horizontal stretch (right neighbor)
            if j < grid_size - 1:
                avg_distance = (u_matrix[i, j] + u_matrix[i, j + 1]) / 2
                x[i, j + 1] = x[i, j] + avg_distance * displacement_factor

            # Vertical stretch (bottom neighbor)
            if i < grid_size - 1:
                avg_distance = (u_matrix[i, j] + u_matrix[i + 1, j]) / 2
                y[i + 1, j] = y[i, j] + avg_distance * displacement_factor

    # Center the grid after applying the stretching
    x -= (x.mean() - x_center)
    y -= (y.mean() - y_center)

    return x, y

def visualize_som_stretched_web_smooth(som, u_matrix):
    grid_size = som.grid_size
    fig, ax = plt.subplots(figsize=(8, 8))

    # Smooth and centered stretching of the grid positions based on the U-matrix values
    x, y = stretch_grid_positions_smooth(grid_size, u_matrix)

    # Flatten the grid for easier iteration
    x_flat = x.flatten()
    y_flat = y.flatten()

    # Plot each neuron as a node in the grid
    scatter = ax.scatter(x_flat, y_flat, s=100, c=u_matrix.flatten(), cmap='viridis', zorder=2)

    # Draw lines between neighboring neurons based on the U-matrix values
    for i in range(grid_size):
        for j in range(grid_size):
            # Plot horizontal connections (right neighbor)
            if j < grid_size - 1:
                ax.plot([x[i, j], x[i, j + 1]], [y[i, j], y[i, j + 1]], c='gray', lw=1, zorder=1)

            # Plot vertical connections (bottom neighbor)
            if i < grid_size - 1:
                ax.plot([x[i, j], x[i + 1, j]], [y[i, j], y[i + 1, j]], c='gray', lw=1, zorder=1)

    # Ensure equal scaling and grid layout
    ax.set_aspect('equal')

    ax.set_title("Net Topology using U Matrix")
    # Label x and y axes
    ax.set_xlabel("X Grid Position")
    ax.set_ylabel("Y Grid Position")

    # Add colorbar
    plt.colorbar(scatter, ax=ax, label='Average Distance to Neighbors')

    # Show the plot
    plt.show()

# Assuming som and u_matrix have been calculated
visualize_som_stretched_web_smooth(som, u_matrix)
```

Los dos graficos representan la misma informacion, es interesante notar que donde hay mas acumulacion de nodos es donde la red tiene mas precision, ya que puede diferencias valores muy similares

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

def plot_variable_on_kohonen_grid(kohonen: Kohonen, variable_matrix, variable_name):
    fig, ax = plt.subplots(figsize=(5, 4))
    plt.title(f'{variable_name} Value For Each Neuron', fontsize=8)
    plt.xlim(-0.5, kohonen.grid_size - 0.5)
    plt.ylim(-0.5, kohonen.grid_size - 0.5)

    norm = mcolors.Normalize(vmin=np.min(variable_matrix), vmax=np.max(variable_matrix))

    for i in range(kohonen.grid_size):
        for j in range(kohonen.grid_size):
            color = plt.cm.magma(norm(variable_matrix[i, j]))  # Color based on variable value
            circle = plt.Circle((j, i), 0.4, color=color, ec='black', lw=0.5)
            ax.add_patch(circle)

    sm = plt.cm.ScalarMappable(cmap='magma', norm=norm)
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax)
    cbar.set_label(variable_name + " Weight Value")


    # Set the ticks to start from 1 and increment by 1 for each position
    num_ticks = u_matrix.shape[0]
    tick_positions = np.arange(num_ticks)
    tick_labels = tick_positions + 1  # Labels from 1 to num_ticks

    plt.xticks(tick_positions, tick_labels)
    plt.yticks(tick_positions, tick_labels)

    plt.xlabel("X Grid Position")
    plt.ylabel("Y Grid Position")

    plt.gca()
    plt.show()

def plot_all_features(kohonen: Kohonen, config):
    feature_names = config['data']['features']

    for i, feature_name in enumerate(feature_names):
        variable_matrix = kohonen.calculate_variable_matrix(feature_column=i)

        plot_variable_on_kohonen_grid(kohonen, variable_matrix, feature_name)


plot_all_features(som, config)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os

def visualize_som_nodes_with_edges_scaled(weights, input_data, grid_size, show=True):
    weights_flat = weights.reshape(-1, weights.shape[-1])  # Flatten for plotting

    feature1 = weights_flat[:, 0]  # First feature dimension
    feature2 = weights_flat[:, 1]  # Second feature dimension

    fig, ax = plt.subplots(figsize=(10, 8))
    ax.scatter(feature1, feature2, s=100, c='blue', alpha=0.7)

    for i in range(grid_size):
        for j in range(grid_size):
            idx = i * grid_size + j
            if i < grid_size - 1:
                ax.plot([feature1[idx], feature1[(i + 1) * grid_size + j]],
                        [feature2[idx], feature2[(i + 1) * grid_size + j]], c='lightgray')
            if j < grid_size - 1:
                ax.plot([feature1[idx], feature1[i * grid_size + (j + 1)]],
                        [feature2[idx], feature2[i * grid_size + (j + 1)]], c='lightgray')

    ax.set_title('Neuron Locations Across Training Iterations')
    ax.set_xlabel('Area')
    ax.set_ylabel('GDP')
    ax.grid(True)
    ax.axis('equal')

    global_min = np.min(input_data, axis=0) - 1
    global_max = np.max(input_data, axis=0)

    # Set consistent limits for x and y axes
    ax.set_xlim(global_min[0], global_max[0])
    ax.set_ylim(global_min[1], global_max[1])

    if show:
        plt.show()
    return fig

def create_som_training_gif(weight_snapshots, input_data, grid_size, output_path="som_training.gif", duration=2, scaled=True):
    frames = []
    for idx, weights in enumerate(weight_snapshots):
        fig = visualize_som_nodes_with_edges_scaled(weights, input_data, grid_size, show=False)

        # Add iteration counter as text overlay
        plt.text(0.05, 1.05, f"Iteration: {idx + 1}", transform=plt.gca().transAxes,
                 fontsize=12, color='blue', fontweight='bold')

        filename = "temp_frame.png"
        fig.savefig(filename)
        plt.close(fig)

        frames.append(imageio.imread(filename))
        os.remove(filename)

    # Save GIF with specified duration between frames
    imageio.mimsave(output_path, frames, fps=2, loop=0)

create_som_training_gif(som.weight_snapshots, som.input_data, som.grid_size, output_path="convergence.gif", scaled=True)
create_som_training_gif(som_2.weight_snapshots, som_2.input_data, som_2.grid_size, output_path="non_convergence.gif", scaled=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set the smoothing window (try larger values if the plot is still too dense)
window_size = 100  # Averaging over 100 epochs

# Convert error lists to pandas Series for easier rolling average calculation
quant_errors_1 = pd.Series(som.quantization_errors)
topo_errors_1 = pd.Series(som.topographic_errors)
quant_errors_2 = pd.Series(som_2.quantization_errors)
topo_errors_2 = pd.Series(som_2.topographic_errors)

# Calculate rolling average to smooth out the lines for both SOMs
smoothed_quant_errors_1 = quant_errors_1.rolling(window=window_size).mean()
smoothed_topo_errors_1 = topo_errors_1.rolling(window=window_size).mean()
smoothed_quant_errors_2 = quant_errors_2.rolling(window=window_size).mean()
smoothed_topo_errors_2 = topo_errors_2.rolling(window=window_size).mean()

# Plotting
plt.figure(figsize=(10, 6))
# Plot errors for som
plt.plot(smoothed_quant_errors_1, label='Quantization Error (SOM 1, Smoothed)', color='blue', linestyle='-')
plt.plot(smoothed_topo_errors_1, label='Topographic Error (SOM 1, Smoothed)', color='blue', linestyle='--')
# Plot errors for som_2
plt.plot(smoothed_quant_errors_2, label='Quantization Error (SOM 2, Smoothed)', color='red', linestyle='-')
plt.plot(smoothed_topo_errors_2, label='Topographic Error (SOM 2, Smoothed)', color='red', linestyle='--')

# Titles and labels
plt.title('Quantization and Topographic Errors (Smoothed) for SOM 1 and SOM 2')
plt.xlabel('Fitted Values')
plt.ylabel('Error')
plt.legend()
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```
